```python
import matplotlib.pyplot as plt
import requests
import time

url = "https://alpha-vantage.p.rapidapi.com/query"

querystrings = [{"from_currency":"XRP", "to_currency":"USD", "function":"CURRENCY_EXCHANGE_RATE"}, 
                {"from_currency":"BTC", "to_currency":"USD", "function":"CURRENCY_EXCHANGE_RATE"},             
                {"from_currency":"ETH", "to_currency":"USD", "function":"CURRENCY_EXCHANGE_RATE"}]

headers = {
    "X-RapidAPI-Key": "83dec6a0a3msh71c744df88f9867p168254jsn31296a7dcfae",
    "X-RapidAPI-Host": "alpha-vantage.p.rapidapi.com"
}

x1, y1, x2, y2, x3, y3 = [], [], [], [], [], []



while True:
    for i in range(len(querystrings)):
        querystring = querystrings[i]
        response = requests.request("GET", url, headers=headers, params=querystring)
        data = response.json()
        # extract latest price and time
        latest_price = float(data["Realtime Currency Exchange Rate"]["5. Exchange Rate"])
        latest_time = data["Realtime Currency Exchange Rate"]["6. Last Refreshed"]
        # datetime
        if i == 0:
            x1.append(latest_time)
            y1.append(latest_price)
            ax1.plot(x1, y1, color='b', label='XRP')
            ax1.set_title('XRP to USD')
        elif i == 1:
            x2.append(latest_time)
            y2.append(latest_price)
            ax2.plot(x2, y2, color='r', label='BTC')
            ax2.set_title('BTC to USD')
        elif i == 2:
            x3.append(latest_time)
            y3.append(latest_price)
            ax3.plot(x3, y3, color='g', label='ETH')
            ax3.set_title('ETH to USD')
  
    
    fig.canvas.draw()
    time.sleep(60)
```
